```python
# Import pipelines
from transformers import pipeline
```

```python
classifier = pipeline("sentiment-analysis")
```

```python
# using classifier
classifier("We are very happy to show you the implementation of Transformers library")
```

```python
text1="fish is delisious in the nearby resturant is poor"
classifier(text1)
```

```python
text_list = ["We are very happy to show you the implementation of Transformers library","fish in the nearby resturant is delicious"]
results = classifier(text_list)

for result in results:
  print(f"label:{result['label']},with_score:{round(result['score'],4)}")
```
